# Informações Dataset
Nome: Acidentes Ferroviários de 2004 a novembro de 2020

Descrição: Registros de acidentes ocorridos nos trechos concedidos às operadoras ferroviárias, juntamente com o processo de monitoramento e análise dos dados no Sistema Acompanhamento de Fiscalização do Transporte Ferroviário 
(SAFF).
 
Registros: 10.824
Colunas: 22


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib  inline
```

```python
# Carregamento do Dataset
df = pd.read_csv("acidentes_ferroviarios_2004-2020.csv", encoding='ISO-8859-1', sep=';')
df.head()
```

```python
# Informações Básicas
df.info()
df.describe()
```

```python
# Média
df.mean(numeric_only=True)
```

```python
# Mediana
df.median(numeric_only=True)
```

```python
# Moda

df.mode().iloc[0]  # primeira moda
```

```python
# Mínimo
df.min(numeric_only=True)
```

```python
# Máximo
df.max(numeric_only=True)
```

```python
# Desvio padrão
df.std(numeric_only=True)
```

```python
# Variância
df.var(numeric_only=True)
```

```python
# Quartis
df.quantile([0.25, 0.5, 0.75], numeric_only=True)
```

```python
# IQR
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
IQR
```

```python
# Valores ausentes
df.isnull().sum()
```

```python
# Histograma
df.hist(figsize=(10,8))
plt.show()
```

```python
# Boxplot
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x=df["Interrupção"], y=df["N_feridos"])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Concessionaria", order=df["Concessionaria"].value_counts().index)
plt.xticks(rotation=45)
plt.ylabel("Número de acidentes")
plt.show()
```

```python
total_feridos = df.groupby("Concessionaria")["N_feridos"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=total_feridos.index, y=total_feridos.values)
plt.xlabel("Total de feridos")
plt.show()
```

```python
df["Natureza"] = df["Natureza"].fillna("Desconhecida")
tipo_de_acidente = df.groupby("Natureza")["N_feridos"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=tipo_de_acidente.index, y=tipo_de_acidente.values)
plt.xlabel("Total de feridos")
plt.show()
```

```python
acidentes_fatais = df.groupby("Natureza")["N_obitos"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=acidentes_fatais.index, y=acidentes_fatais.values)
plt.xlabel("Total de feridos")
plt.show()
```

```python
import pandas as pd
```

```python
# Carregar csvs
countries_df = pd.read_csv("countries.csv")
countries_df.info()
```

```python
flags_df = pd.read_csv("flags.csv")
flags_df.info()
```

```python
# Ver interseção
intersecao = pd.merge(countries_df, flags_df, on="name")
intersecao.info()
```

```python
somente_em_flags = flags_df[flags_df.name.isin(countries_df.name) == False]
somente_em_flags.info()
```

```python
somente_em_countries = countries_df[countries_df.name.isin(flags_df.name) == False]
somente_em_countries.info()
```
